```python
% matplotlib inline
import matplotlib.pyplot as plt 

x = [7,5,7,3,4,1,0,2,8,6,5,3]
y = [5,7,7,3,6,4,0,2,7,8,5,7]
# create new figure
fig = plt.figure(figsize=(12,8))

# divide subplots into 2 x 3 grid
# and select #1
ax_1 =fig.add_subplot(231)
ax_1.plot(x, y,linewidth=3.0)
ax_1.set_xlabel("Features")
ax_1.set_ylabel("Predictor")
ax_1.set_title("Simple Plot")

# select #2
ax_2 =fig.add_subplot(232)
ax_2.bar(x, y)
ax_2.set_title("Bar Plot")

ax_3=fig.add_subplot(233)
ax_3.barh(x, y)

ax_4 =fig.add_subplot(234)
ax_4.bar(x, y)

# box plot
ax_5=fig.add_subplot(235)
ax_5.boxplot(x)

# scatter plot
ax_6=fig.add_subplot(236)
ax_6.scatter(x,y)
```
