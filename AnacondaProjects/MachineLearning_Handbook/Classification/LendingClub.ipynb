```python
import os.path
import datetime
import re
import numpy as np
import pandas as pd
import seaborn as sns
import warnings

from mlsettings.settings import load_app_config, get_datafolder_path
from mltools.modelbuilder.supervised  import SupervisedDataLoader 

warnings.filterwarnings('ignore')
% matplotlib inline 
np.set_printoptions(precision=4)

pd.set_option('display.width', 200)
pd.set_option('precision', 4)
pd.set_option('display.width', 500)
pd.set_option('display.max_columns', 100)
pd.set_option('display.notebook_repr_html', True)
sns.set_style("whitegrid")
pd.options.display.float_format = '{:,.4f}'.format
sns.set()
import logging
logger = logging.getLogger()
formatter = logging.Formatter('%(asctime)s - %(name)s - %(levelname)s - %(message)s', datefmt='%d-%b-%y %H:%M:%S')
logger.setLevel(logging.DEBUG)
```

```python
load_app_config()
DIRECTORY="lendingclub"
TRAIN_FILE  = "Lending_Train.csv"
TEST_FILE  = "Lending_Test.csv"
RESPONSE = "loan_status"
input_path = get_datafolder_path()
train_file = os.path.join(input_path, DIRECTORY, TRAIN_FILE)
test_file = os.path.join(input_path, DIRECTORY, TEST_FILE)
supervisedloader = SupervisedDataLoader(train_file=train_file,test_file=test_file,response =RESPONSE) 
train_dataset,test_dataset = supervisedloader.load()
supervisedloader.response
```

```python
train_dataset[RESPONSE].head()
```

```python
from datatools.customtransformers import LabelTransformer
label_transformer  = LabelTransformer([RESPONSE])
X_train = label_transformer.fit_transform(train_dataset)
X_train.head()
```

```python
y_train = X_train[RESPONSE]
X_train = X_train.drop(RESPONSE,axis =1)
X_test  = test_dataset.drop(RESPONSE,axis=1)
```

```python
y_train.value_counts()
```

```python
train_dataset[RESPONSE].value_counts()
```

```python
X_train.head()
```

```python
from mltools.mlhelper import missing_values
X_train_missing = missing_values(X_train)
X_test_missing = missing_values(X_test)
X_train_length = X_train.shape[0]
X_test_length = X_test.shape[0]
missing_training_all  = X_train_missing[X_train_missing['Missing_Values'] == X_train_length]
missing_testing_all  = X_test_missing[X_test_missing['Missing_Values'] == X_test_length]
print("No of columns that have 100% missing values in training dataset :{0}".format(missing_training_all.shape[0]))
print("No of columns that have 100% missing values in test dataset :{0}".format(missing_testing_all.shape[0]))
```

```python
X_train_missing.head()
```

```python
features_to_drop = missing_testing_all['Feature'].tolist()
```

```python
X_train = X_train.drop(features_to_drop,axis =1)
X_test  = X_test.drop(features_to_drop,axis =1)
X_train.shape
```

```python
feature_count  = X_train.nunique(dropna=False)
constant_features = pd.DataFrame({'feature':feature_count.index,'count':feature_count.values})
print("constant features :{0},{1}".format(*constant_features.shape))
remove_constants =constant_features[constant_features['count']==1]['feature'].tolist()
remove_constants
```

```python
print("Before removing constant features :{0}".format(X_train.shape))
X_train.drop(remove_constants,axis =1,inplace=True)
print("After removing constant features :{0}".format(X_train.shape))
```

```python
print("Before removing constant features :{0}".format(X_test.shape))
X_test.drop(remove_constants,axis =1,inplace=True)
print("After removing constant features :{0}".format(X_test.shape))
```

```python
X_train.head()
```

```python
def get_nullframe(dataset):
    null_sum = dataset.isnull().sum()[dataset.isnull().sum()>0] 
    return pd.DataFrame({'feature':null_sum.index,'count':null_sum.values,
                          'null_percent':(null_sum.values/dataset.shape[0])*100}).sort_values(by=['count'],ascending=False) 
```

```python
X_train_null =get_nullframe(X_train)
X_train_null
```

```python
X_test_null =get_nullframe(X_test)
```

```python
X_train['term'] =X_train['term'].apply(lambda x:x.replace('months','')).astype(int)
X_test['term'] =X_test['term'].apply(lambda x:x.replace('months','')).astype(int)
```

```python
X_train['int_rate'] =X_train['int_rate'].apply(lambda x:x.replace('%','')).astype(float)
X_test['int_rate'] =X_test['int_rate'].apply(lambda x:x.replace('%','')).astype(float)
```

```python
X_test['emp_length'] =X_test['emp_length'].fillna('0')
X_train['emp_length'] =X_train['emp_length'].fillna('0')
X_train['emp_length'] =X_train['emp_length'].apply(lambda x:re.sub(r'years?|\<|\+| ','',x )).astype(int)
X_test['emp_length'] =X_test['emp_length'].apply(lambda x:re.sub(r'years?|\<|\+| ','',x )).astype(int)
```
