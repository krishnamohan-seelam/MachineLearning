### Step 1. Import the necessary libraries

```python
import pandas as pd
import collections
import matplotlib.pyplot as plt 

# set this so the graphs open internally
%matplotlib inline
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv). 

### Step 3. Assign it to a variable called chipo.

```python
filename  = r'data\chipotle.tsv'
chipo = pd.read_csv(filename,sep ='\t')
```

### Step 4. See the first 10 entries

```python
plt.style.use('ggplot')
chipo.head(10)
```

### Step 5. Create a histogram of the top 5 items bought

```python
top_five = chipo['item_name'].value_counts().nlargest(5)
fig = plt.figure(figsize=(8,6))
ax = fig.gca()
top_five.plot(kind='bar',ax=ax,rot=45)
ax.set_xticklabels(ax.get_xticklabels(),ha='right')
plt.show()
```

### Step 6. Create a scatterplot with the number of items orderered per order price
#### Hint: Price should be in the X-axis and Items ordered in the Y-axis

```python
convert_float =lambda x:float(x[1:-1])
item_price_ds =chipo.item_price.apply(convert_float) 
chipo['item_price'] = item_price_ds
```

```python
 
fig = plt.figure(figsize=(8,6))
ax = fig.gca()
chipo.plot(x='item_price',y ='quantity',ax=ax,kind='scatter')
```

### Step 7. BONUS: Create a question and a graph to answer your own question.

```python
least_five = chipo['item_name'].value_counts().nsmallest(5)
fig = plt.figure(figsize=(8,6))
ax = fig.gca()
least_five.plot(kind='bar',ax=ax,rot= 45)
ax.set_xticklabels(ax.get_xticklabels(),ha='right')
plt.show()
```
