```python
import numpy as np
import pandas as pd
```

```python
#Create a DataFrame
d = {
'Name':['Alisa','Bobby','Cathrine','Alisa','Bobby','Cathrine','Alisa','Bobby','Cathrine','Alisa','Bobby','Cathrine'],
'Class':[6,6,6,5,5,5,4,4,4,3,3,3]    ,
'Subject':['Mathematics','Mathematics','Mathematics','Science','Science','Science',
'History','History','History','Economics','Economics','Economics'],
'Score':[62,47,55,74,31,77,85,63,42,62,89,85]}
 
df = pd.DataFrame(d,columns=['Name','Class','Subject','Score'])
df
```

```python
sorted_df = df.sort_values(by=['Class','Subject','Score'],ascending=False)
sorted_df
```

```python
#sorted_df.nlargest(1,['Subject','Score'],keep='first')
df["group_rank"] = df.groupby(["Class","Subject"])["Score"].rank(ascending=0,method='dense')
df.sort_values(by=["group_rank"],ascending=True)
```

```python
df[df["group_rank"]==1]
```

```python
import datetime
sample_data = {
    'id':[1,1,2,2,3,3,4,4],
    'submit_date':[datetime.date(year=2018,month =10,day=5),datetime.date(year=2018,month =10,day=3),
                   datetime.date(year=2018,month =4,day=3),datetime.date(year=2018,month =2,day=1),
                   datetime.date(year=2018,month =6,day=5),datetime.date(year=2018,month =1,day=5),
                   datetime.date(year=2018,month =1,day=5),datetime.date(year=2018,month =1,day=5),
                  ],
    'amount':[100,0,200,0,100,0,100,0]
}
sample = pd.DataFrame(sample_data,columns=['id','submit_date','amount'])
sample
```

```python
sample_sort = sample.sort_values(['id','submit_date','amount'], ascending=False)
sample_sort
```

```python
sample_sort['rn']=sample_sort.groupby(['id']).cumcount()+1
sample_sort 
```

```python
sample_sort[sample_sort['rn']==1]
```
