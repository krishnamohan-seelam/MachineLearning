```python
import pathlib
```

```python
import pandas as pd
```

```python
cwd = pathlib.Path().cwd()
```

```python
data_folder = cwd.joinpath('data')
```

```python
inp_file = data_folder.joinpath('legislators-historical.csv')
```

```python
dtypes = {
    "first_name": "category",
    "gender": "category",
    "type": "category",
    "state": "category",
    "party": "category",
}
usecols=list(dtypes) + ["birthday", "last_name"]
```

```python
df = pd.read_csv(inp_file,dtype = dtypes,usecols=usecols)
```

```python
df.head()
```

```python
groupby_state = df.groupby(['state'])['last_name'].aggregate(namecount =('last_name','count')).reset_index()
```

```python
groupby_state.head()
```

```python
df.groupby(["state", "gender"]).agg(name_size = pd.NamedAgg(column='last_name',aggfunc='size')).reset_index()
```
