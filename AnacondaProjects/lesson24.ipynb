```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
mu ,sigma  =  0, 0.1
s = np.random.normal(mu,sigma,1000)


print("Verify the mean:",abs(mu - np.mean(s)) < 0.01)

print("Verify the variance:",abs(sigma - np.std(s, ddof=1)) < 0.01)

count, bins, ignored = plt.hist(s, 30, normed=True)
#probability density for the Gaussian distribution

plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) *
               np.exp( - (bins - mu)**2 / (2 * sigma**2) ),
         linewidth=2, color='r')
                         
```
