```python
from bokeh.plotting  import figure
```

```python
from bokeh.io import output_file,show,output_notebook
```

```python
x = list(range(0,6))
y = [i**2 for i in x]
#output_file("line.html")
output_notebook()
fig = figure(plot_width=400, plot_height=300)
fig.line(x,y)
show(fig)
```

```python
p = figure(plot_width=400, plot_height=300)

# add a circle renderer with a size, color, and alpha
p.circle(x,y, size=10, color="navy", alpha=0.5)

# show the results
show(p)
```

```python
bachelors_file = r'data/bachelors.csv'
bachelors_df = pd.read_csv(bachelors_file)
bachelors_df.head()
```

```python
fig = figure(plot_width =400,plot_height =300)
fig.xaxis.axis_label = 'Year'
fig.yaxis.axis_label = 'Engineering'
fig.line(bachelors_df['Year'],bachelors_df['Engineering'])
show(fig)
```

```python
p=figure(plot_width=500,plot_height=400, tools='pan')
 
p.title.text="Cool Data"
p.title.text_color="Gray"
p.title.text_font="lucida"
p.title.text_font_style="bold"
p.xaxis.minor_tick_line_color=None
p.yaxis.minor_tick_line_color=None
p.xaxis.axis_label="Date"
p.yaxis.axis_label="Intensity"    
 
p.line([1,2,3],[4,5,6])
show(p)
```

```python
excel_file = r'data/verlegenhuken.xlsx'
sheet_name ='Ark1'
weather_df = pd.read_excel(excel_file,sheet_name=sheet_name)
weather_df.head()
```

```python
fig = figure(plot_width =400, plot_height=300)
fig.title.text="Temperature and Air Pressure"
fig.title.text_color="Gray"
fig.title.text_font="lucida"
fig.title.text_font_style="bold"
fig.xaxis.minor_tick_line_color=None
fig.yaxis.minor_tick_line_color=None
fig.xaxis.axis_label="Temperature(C)"
fig.yaxis.axis_label="Pressure(hPa)"    
fig.circle(weather_df['Temperature']/10,weather_df['Pressure'], size=0.5 ,alpha=0.5)
show(fig)
```
