```python
datapath = r"data/supermarkets.json"
```

```python
supermarkets =pd.read_json(datapath)
```

```python
supermarkets
```

```python
def full_address(row):
    return row['Address']+', ' +\
           row['City']   +', ' +\
           row['State']  +', ' +\
           row['Country']  
    
```

```python
supermarkets['Full_Address'] = supermarkets.apply(full_address,axis=1)
```

```python
supermarkets
```

```python
supermarkets['location'] = supermarkets['Full_Address'].apply(nom.geocode)
```

```python
supermarkets['latitude'] = supermarkets['location'].apply(lambda x : x.latitude if x else None)
```

```python
supermarkets['longitude'] = supermarkets['location'].apply(lambda x : x.longitude if x else None)
```
