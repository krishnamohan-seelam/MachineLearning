```python
import datetime
import pandas as pd
from bokeh.plotting  import figure
from bokeh.io import output_file,show,output_notebook
output_notebook()
```

```python
data_file = r'data\NASDAQ.csv'
stocks_df= pd.read_csv(data_file)
```

```python
stocks_df.set_index('Date',inplace=True)  
stocks_df.head()
```

```python
hours_12 = 12 *60*60 *1000
fig = figure(plot_width=800, plot_height=300)
fig.rect(stocks_df, (stocks_df['Open']+stocks_df['Close'])/2 ,hours_12,abs(stocks_df['Open']-stocks_df['Close']))
show(fig)
```
